Importing dataset

```python
adult = pd.read_csv('adult.csv')
adult.head()
```

Seperating the dataset into colums having string values and columns having int values

```python
a = adult.iloc[:,[1,3,5,6,7,8,9,13]]
b = adult.iloc[:,[0,2,4,10,11,12,14]]
a
```

```python
a.workclass=a.workclass.astype('category').cat.codes
a.education=a.education.astype('category').cat.codes
a['marital-status']=a['marital-status'].astype('category').cat.codes
a.occupation=a.occupation.astype('category').cat.codes
a.relationship=a.relationship.astype('category').cat.codes
a.race=a.race.astype('category').cat.codes
a.gender=a.gender.astype('category').cat.codes
a['native-country']=a['native-country'].astype('category').cat.codes
```

```python
a
```

```python
import pandas as pd
```

```python
c1 = pd.cut(b['age'], bins=19)
c2 = pd.cut(b['fnlwgt'], bins=19)
c3 = pd.cut(b['educational-num'], bins=19)
c4 = pd.cut(b['capital-gain'], bins=19)
c5 = pd.cut(b['capital-loss'], bins=19)
c6 = pd.cut(b['hours-per-week'], bins=19)
```

```python
b
```

```python
b.income=b.income.astype('category').cat.codes
```

```python
l1 = pd.get_dummies(c1, dtype=int)
l2 = pd.get_dummies(c2, dtype=int)
l3 = pd.get_dummies(c3, dtype=int)
l4 = pd.get_dummies(c4, dtype=int)
l5 = pd.get_dummies(c5, dtype=int)
l6 = pd.get_dummies(c6, dtype=int)
```

```python
b
```

Using One Hot Encoding to encode the string values into integer values

```python
# a_ohe = pd.get_dummies(a)
# a_ohe.shape, b.shape
```

```python
a.shape
```

```python
c = pd.concat([a, l1, l2, l3, l4, l5, l6], axis = 1)
```

```python
c.head()
```

```python
y = b['income']
X = c


data = pd.concat([y, X], axis=1)
data.to_csv("adultset.dat", sep = " ", header=None, index=None, index_label='str')

data = data.to_numpy()
```
